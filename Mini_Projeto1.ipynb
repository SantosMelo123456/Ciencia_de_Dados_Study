```python
#Em vez de usar apenas .sum(), usar .agg() para aplicar *funções* diferentes a colunas diferentes e renomeá-las simultaneamente;

df.groupby('Student_ID').agg(
    Study_Hours_Per_Day= ('Study_Hours_Per_Day','sum'), # vai realisar a função de somar
    Age=('Age', 'mean'),                                # vai realisar a função média
    Sleep_Hours=('Sleep_Hours', 'count')                # vai realisar a função de contar
    
    
)
```

```python
# POR FIM:

# Perguntas principais a fazer a partir do dataset:
# 1. Qual curso têm mais desempenho de horas de estudo por dia? 
# 2. Qual curso têm a media geral (gpa) mais alta? 
# 3. Qual gênero predomina na grande maioria dos cursos?
```

```python
# 1. RESPOSTA:
# criando uma variável  que filtra o cruso com a maior média de horas de estudo.

curso_vencedor = df.groupby('Major')['Study_Hours_Per_Day'].mean().idxmax()
print(f'O curso vencedor com a maior média de horas de estudo foi: {curso_vencedor}')

# Mas qual foi o valor da média máxima? Aqui a resposta:
media_maxima = df.groupby('Major')['Study_Hours_Per_Day'].mean().max()
print(f"O valor da maior média do curso com o maior tempo de estudo foi: {media_maxima}")
```

```python
# 2. RESPOSTA:

maior_gpa = df.groupby('Major')['Final_CGPA'].mean().idxmax()
print(f'O curso com a média de nota final do GPA mais alta foi: {maior_gpa}')

# Tá, mais qual o valor final desse cálculo da média mais alta? Aqui:
maior_media = df.groupby('Major')['Final_CGPA'].mean().max()
print(f"O valor da media mais alta foi de: {maior_media}")
```

```python
# 3. RESPOSTA:

genero_por_curso = df.groupby(['Major', 'Gender']).size().unstack()  # Unstack é como "desempilhar" um nível de linhas (index) e transformá-lo em colunas.
print(genero_por_curso)

# Qual gênero vence em mais cursos?
genero_dominante = genero_por_curso.idxmax(axis=1)                   # axis significa: indica a direção ao longo da qual uma operação será aplicada. axis=1 → opera ao longo das colunas (ou seja, a função é aplicada por linha). Ou seja, "atravessa" pelas colunas (horizontal).

print("\nGênero dominante por curso:")
print(genero_dominante)

# Vencedor final:
vencedor_geral = genero_dominante.value_counts().idxmax()            # Values_count significa: o resultado vem ordenado em ordem decrescente de frequência.
print(f"\nNo geral, o gênero predominante na maioria dos cursos é: {vencedor_geral}")
```
